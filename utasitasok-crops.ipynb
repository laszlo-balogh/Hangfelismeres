```python
dirs = ["ajto_bezar-spec","klima_le-spec","mennyi_az_ido-spec","tv_be-spec","vicc-spec","villany_fel-spec","villany_le-spec"]
```

```python

filenames = []
for dir in dirs:
    #globals()[f"{dir}"]=[]
    file_list=next(walk(f"utasitasok-melspec/{dir}"), (None, None, []))[2]
    for x in file_list:
       filenames.append(x)



```

```python
ajto_be_index = 1;
klima_le_index = 1;
ido_index = 1;
tv_be_index = 1;
vicc_index = 1;
villany_fel_index = 1;
villany_le_index = 1;

for x in filenames:
    dir = x.split('-')[0] #ajto_bezar | -1.png
    filename = f"utasitasok-melspec/{dir}/{x}"    
    if dir == "ajto_bezar": #1
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{ajto_be_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{ajto_be_index}.png", crop_image)
        ajto_be_index = ajto_be_index+1
    elif dir == "klima_le": #2
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{klima_le_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{klima_le_index}.png", crop_image)
        klima_le_index = klima_le_index+1
    elif dir == "mennyi_az_ido": #3
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{ido_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{ido_index}.png", crop_image)
        ido_index = ido_index+1
    elif dir == "tv_be": #4
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{tv_be_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{tv_be_index}.png", crop_image)
        tv_be_index = tv_be_index+1
    elif dir == "vicc": #5
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{vicc_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{vicc_index}.png", crop_image)
        vicc_index = vicc_index+1
    elif dir == "villany_fel": #6
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{villany_fel_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{villany_fel_index}.png", crop_image)
        villany_fel_index = villany_fel_index+1
    else:
        image = cv2.imread(f'utasitasok-melspec/{dir}-spec/{dir}-{villany_le_index}.png')
        crop_image = image[ 58:428,80:478]
        cv2.imwrite(f"utasitasok-spec-crop/{dir}-crop/{dir}-cropped-{villany_le_index}.png", crop_image)
        villany_le_index = villany_le_index+1
```
